# Fit to generated data
I will fit to sec1_2 data both via scipy functions as well as through the method I will discuss on the website

```python
import numpy as np
from scipy.linalg import lstsq, inv
from matplotlib import pylab as plt
```

# Load Data

```python
data=np.loadtxt('sec1_2_data.csv',delimiter=',',skiprows=1)
x=data[:,0]
y=data[:,1]
```

# Scipy

```python
M = x[:, np.newaxis]**[0, 1]

p, res, rnk, s = lstsq(M, y)

plt.plot(x, y, 'o', label='data')

xx = np.linspace(min(x), max(x), 300)

yy = p[0] + p[1]*xx

plt.plot(xx, yy, label='least squares fit, $y = a + bx$')

plt.xlabel('x')

plt.ylabel('y')

plt.legend(framealpha=1, shadow=True)

# plt.grid(alpha=0.25)

plt.grid(True)

plt.show()
```

# Manual Fit

```python
npoints = len(x)
# print(npoints)
A = np.vstack((np.ones(npoints,),x)).T
paramfit = inv(A.T @ A) @ A.T @ y

print(paramfit)
res = y - A @ paramfit
yhat = A @ paramfit
print(res.T @ res)

plt.plot(x, y, 'o', label='Data')
plt.plot(x, yhat, label='Least squares fit, $y(t) = x_1 + x_2t$')

plt.xlabel('t')

plt.ylabel('y')

plt.legend(framealpha=1, shadow=True)

# plt.grid(alpha=0.25)

plt.grid(True)
plt.savefig('myfit.png', bbox_inches='tight')
```
